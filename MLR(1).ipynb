```python
startup = startup.rename({'R&D Spend':'rd_spend','Marketing Spend':'marketing_spend'},axis=1)
startup
```

```python
import statsmodels.formula.api as smf
model=smf.ols('Profit~rd_spend+Administration+marketing_spend',data=startup).fit()
```

```python
model.summary()
```

```python
# coefficients 
model.params
```

```python
#t and p-values
print(model.tvalues, '\n',model.pvalues)
```

```python
(model.rsquared,model.rsquared_adj)
```

```python
ml_v=smf.ols('Profit~marketing_spend ',data=startup).fit()
print(ml_v.tvalues, '\n',ml_v.pvalues)
```

```python
ml_w=smf.ols('Profit~rd_spend',data=startup).fit()
print(ml_w.tvalues, '\n',model.pvalues)
```

```python
ml_wv=smf.ols('Profit~rd_spend+Administration',data=startup).fit()
print(ml_wv.tvalues, '\n',model.pvalues)
```

```python
rsq_rd = smf.ols('rd_spend~marketing_spend+Administration',data=startup).fit().rsquared  
vif_rd = 1/(1-rsq_rd) 

rsq_A = smf.ols('Administration~rd_spend+marketing_spend',data=startup).fit().rsquared  
vif_A = 1/(1-rsq_A) 

rsq_m = smf.ols('marketing_spend~Administration+rd_spend',data=startup).fit().rsquared  
vif_m = 1/(1-rsq_m) 
d1 = {'Variables':['Administration','rd_spend','marketing_spend'],'VIF':[vif_A,vif_rd,vif_m]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

```python
import statsmodels.api as sm
qqpplot=sm.qqplot(model.resid,line='q') 
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
def get_standerdised_value( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standerdised_value(model.fittedvalues),
            get_standerdised_value(model.resid))

plt.title('residual plot')
plt.xlabel('model.fitted values')
plt.ylabel('model.resid')
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model,"rd_spend", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8 ))
fig = sm.graphics.plot_regress_exog(model, "Administration" , fig =fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "marketing_spend" , fig=fig)
plt.show()
```

```python
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
# plot the influencers plot using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(startup)), np.round(c, 3))
plt.xlabel('rows index')
plt.ylabel('cooks Distance')
plt.show()
```

```python
(np.argmax(c),np.max(c))
```

```python
# index and value of influencer wher c is more than .5
influence_plot(model)
plt.show()
```
